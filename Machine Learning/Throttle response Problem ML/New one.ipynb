```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import pickle
```

```python
throttle_data = pd.read_csv('live1.csv')
```

```python
dropped_dataset = throttle_data.drop(columns=['LONG_TERM_FUEL_TRIM_BANK_1 ()','SHORT_TERM_FUEL_TRIM_BANK_1 ()','INTAKE_MANIFOLD_PRESSURE ()','FUEL_TANK ()','ABSOLUTE_THROTTLE_B ()','PEDAL_D ()','PEDAL_E ()','COMMANDED_THROTTLE_ACTUATOR ()','ABSOLUTE_BAROMETRIC_PRESSURE ()','RELATIVE_THROTTLE_POSITION ()','TIMING_ADVANCE ()','CATALYST_TEMPERATURE_BANK1_SENSOR1 ()','CATALYST_TEMPERATURE_BANK1_SENSOR2 ()','CONTROL_MODULE_VOLTAGE ()','COMMANDED_EVAPORATIVE_PURGE ()','TIME_RUN_WITH_MIL_ON ()','TIME_SINCE_TROUBLE_CODES_CLEARED ()','DISTANCE_TRAVELED_WITH_MIL_ON ()','WARM_UPS_SINCE_CODES_CLEARED ()'])
dropped_dataset
```

```python
throttle_threshold = 20
dropped_dataset['Throttle_Response_Problem'].fillna((dropped_dataset['THROTTLE ()']> throttle_threshold).astype(int),inplace=True)
dropped_dataset.to_csv('updated_livenew.csv',index=False)
updated_throttle_data  = pd.read_csv('updated_live1.csv')
updated_throttle_data
```

```python
features = ['ENGINE_RUN_TINE ()','ENGINE_RPM ()','VEHICLE_SPEED ()','THROTTLE ()','ENGINE_LOAD ()','COOLANT_TEMPERATURE ()','FUEL_AIR_COMMANDED_EQUIV_RATIO ()','INTAKE_AIR_TEMP ()']
target = 'Throttle_Response_Problem'
```

```python
x = updated_throttle_data[features]
y = updated_throttle_data[target]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) 
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
clf = RandomForestClassifier(random_state=42)
clf.fit(x_train_scaled, y_train)
```

```python
pickle.dump(clf, open("throttle.pkl", "wb"))
pickle.dump(scaler, open("scaler.pkl", "wb"))
```

```python
y_pred = clf.predict(x_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
```

```python
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
